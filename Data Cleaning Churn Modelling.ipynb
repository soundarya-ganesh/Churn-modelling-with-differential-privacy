```python
def column_details(col):
    print("The column name is: ", col)
    print("% NULL VALUES: ", "%", round(df[col].isnull().sum()/df.shape[0]*100, 2))
    print("# NULL VALUES: ", df[col].isnull().sum())
    print("Unique Counts: ", df[col].nunique())
    print(df[col].value_counts(dropna = False))
```

```python
def clean(df):
    # Drop irelative features
    result = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1)
    
    # One Hot Encoding
    result = pd.get_dummies(result, columns = ['Geography', 'Gender'], drop_first = True)
    
    return result
```

```python
new_df = clean(df)
```
